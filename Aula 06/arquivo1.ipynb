```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
imagem = cv2.imread('bentoespock.jpeg')
```

```python
type(imagem)
```

```python
plt.imshow(imagem)
plt.show()
```

```python
nova_imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(nova_imagem)
plt.show()
```

```python
nova_imagem.shape
```

```python
redimensionada_imagem = cv2.resize(nova_imagem, (640,480))
```

```python
plt.imshow(redimensionada_imagem)
plt.show()
```

```python
redimensionada_imagem.shape
```

```python
largura = 0.1
altura = 0.1
```

```python
outra_imagem = cv2.resize(nova_imagem, (0,0), nova_imagem, largura, altura)
```

```python
plt.imshow(outra_imagem)
plt.show()
```

```python
outra_imagem.shape
```

```python
imagem_girada = cv2.flip(nova_imagem, 0)
```

```python
plt.imshow(imagem_girada)
plt.show()
```

```python
cv2.imwrite('bento_modificado.jpg', imagem)
```

```python
cv2.imwrite('bento_girado.jpg', imagem_girada)
```

```python
img = plt.figure(figsize=(20,10))
ax = img.add_subplot(111)
ax.imshow(imagem_girada)
plt.show()
```
