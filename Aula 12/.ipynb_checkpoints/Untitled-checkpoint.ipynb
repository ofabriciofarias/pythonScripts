```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 
          'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:

    #Criar uma copia da imagem
    copia_imagem = imagem.copy()
    
    method = eval(m)
    
    #Templat Matching
    res = cv2.matchTemplate(copia_imagem, rosto_imagem, method)
    
    #Depois pegamos os valores e localizaes minimas e mximas para 
    #desenhar o retangulo
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc #(x,y)
    else:
        top_letf = max_loc
        
        
    #definio da forma do retngulo
    height, width, channels = rosto_imagem.shape
    
    botton_right = (top_left[0] + width, top_left[1] + height)
    
    cv2.rectangle(imagem, top_left, botton_right, (255,0,0), 15)
    
    #Apresentar a imagem
    plt.subplot(121)
    
    plt.subplot(122)
    
```
