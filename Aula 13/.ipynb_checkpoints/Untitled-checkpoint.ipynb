```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xadrez = cv2.imread('xadrez.jpg')
xadrez = cv2.cvtColor(xadrez, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(xadrez)
plt.show()
```

```python
cinza_xadrez = cv2.cvtColor(xadrez, cv2.COLOR_RGB2GRAY)
plt.imshow(cinza_xadrez, cmap='gray')
#plt.axis('off')
plt.show()
```

```python
xadrez_real = cv2.imread('xadrez_real.jpg')
xadrez_real = cv2.cvtColor(xadrez_real, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(xadrez_real)
plt.show()
```

```python
cinza_xadrez_real = cv2.cvtColor(
    xadrez_real, cv2.COLOR_RGB2GRAY)
plt.imshow(cinza_xadrez_real, cmap = 'gray')
plt.show()
```

```python
cinza_xadrez
```

```python
cinza = np.float32(cinza_xadrez)
```

```python
destino = cv2.cornerHarris(
    src = cinza_xadrez, blockSize = 2, ksize = 3, k = 0.04)
```

```python
#transformao morfolgico que dar intensidade na regio de borda
destinho = cv2.dilate(destino, None)
```

```python
#Sempre que tivermos uma border maior que 1% trocamos a cor para vermelho
xadrez[destino>0.01*destinho.max()] = [255,0,0]
```

```python
plt.imshow(xadrez)
plt.show()
```

```python
#Aplicando em uma imagem real
cinza = np.float32(cinza_xadrez_real)
destino = cv2.cornerHarris(
    src = cinza, blockSize = 2, ksize = 3, k = 0.04)
```

```python
destino = cv2.dilate(destino,None)
```

```python
xadrez_real[destino>0.01*destino.max()] = [255,0,0]
```

```python
plt.imshow(xadrez_real)
plt.show()
```
