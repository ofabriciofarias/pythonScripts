```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xadrez = cv2.imread('xadrez.jpg')
xadrez = cv2.cvtColor(xadrez, cv2.COLOR_BGR2RGB)
cinza_xadrez = cv2.cvtColor(xadrez, cv2.COLOR_RGB2GRAY)
plt.imshow(xadrez)
plt.show()
```

```python
bordas = cv2.goodFeaturesToTrack(cinza_xadrez, 48, 0.01, 10)
```

```python
bordas
```

```python
bordas = np.intp(bordas)
```

```python
bordas
```

```python
for i in bordas:
    x,y = i.ravel()
    
    cv2.circle(xadrez, (x,y), 3, (0,255,0), -1)
    
plt.imshow(xadrez)
plt.show()
```

```python
xadrez_real = cv2.imread('xadrez_real.jpg')
xadrez_real = cv2.cvtColor(xadrez_real, cv2.COLOR_BGR2RGB)
cinza_xadrez_real = cv2.cvtColor(xadrez_real, cv2.COLOR_RGB2GRAY)
plt.imshow(xadrez_real)
```

```python
bordas = cv2.goodFeaturesToTrack(cinza_xadrez_real, 80, 0.01, 10)
```

```python
bordas = np.intp(bordas)
```

```python
bordas.shape
```

```python
for i in bordas:
    x,y = i.ravel()
    
    cv2.circle(xadrez_real, (x,y), 3, (0,255,0), -1)
```
