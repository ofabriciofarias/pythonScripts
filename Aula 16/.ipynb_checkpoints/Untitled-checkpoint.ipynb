```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
barcelona = cv2.imread('barcelona.jpg',0)
turma = cv2.imread('turma.jpg', 0)
#apenaseu = cv2.imread('apenaseu.jpg',0)
```

```python
plt.imshow(barcelona, cmap='gray')
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_detector(imagem):
    cont = 0
    face_imagem = imagem.copy()
    
    face_rectangulo = face_cascade.detectMultiScale(face_imagem, scaleFactor = 1.2, minNeighbors = 5)
    
    for(x,y,w,h) in face_rectangulo:
        cv2.rectangle(face_imagem, (x,y), (x+w, y+h), (255,255,255), 10)
        cont = cont +1
        
    print(cont)
    return face_imagem
```

```python
resultado = face_detector(barcelona)
plt.imshow(resultado, cmap = 'gray')
plt.show()
```

```python
cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read(0)
    
    frame = face_detector(frame)
    
    cv2.imshow('Deteco Facial', frame)
    #plt.show()
    
    k = cv2.waitKey(1)
    
    #if k == 27:
     #   break
        
    if cv2.waitKey(1) & 0xFF == ord('q'):
       break
        
cap.release()
cv2.destroyAllWindows()
```
