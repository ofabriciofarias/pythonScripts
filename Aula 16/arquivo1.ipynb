```python
face_cascade = cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_detector(imagem):
    cont = 0
    face_imagem = imagem.copy()
    
    face_retangulo = face_cascade.detectMultiScale(face_imagem, scaleFactor = 1.2, minNeighbors = 5)
    
    for(x,y,w,h) in face_retangulo:
        cv2.rectangle(face_imagem, (x,y), (x+w, y+h), (255,0,0), 10)
        cont = cont + 1
        
    print(cont)
    return face_imagem
```

```python
resultado = face_detector(barcelona)
plt.imshow(resultado, cmap = 'gray')
plt.show()
```
